```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.randint(0, 50, 100)

# Positive correlation with some noise
y = x + np.random.normal(1, 12, 100)
```

```python
np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.randint(0, 50, 100)
y = np.random.randint(0, 50, 100)

np.corrcoef(x, y)

plt.scatter(x,y);
```
