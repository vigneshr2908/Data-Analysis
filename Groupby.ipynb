```python
fortune = pd.read_csv(r'D:\Learning & Projects\Python\Pandas\Boris\fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head(3)
```

```python
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_rev_comp = data.nlargest(2, 'Revenue')
    df = df.append(highest_rev_comp)
```

```python
cities = fortune.groupby('Location')
```

```python
cities.size()
```

```python
for city, data in cities:
    highest_profit_comp = data.nlargest(1, 'Profits')
    df = df.append(highest_profit_comp)
```
