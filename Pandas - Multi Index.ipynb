### Change Index level name with 'set_names' method

```python
bigmac = pd.read_csv(r'D:\Learning & Projects\Python\Pandas\Boris\bigmac.csv', parse_dates=['Date'], 
                     index_col = ['Date', 'Country'])
bigmac.sort_index(inplace=True)
bigmac.head()
```

```python
bigmac.head()
```

```python
bigmac.index.set_names(names = 'Date', level=0, inplace=True)
```

```python
bigmac.sort_index(ascending=False)
bigmac.sort_index(ascending=[True, False])
```

### The Stack method

```python
world = pd.read_csv(r'D:\Learning & Projects\Python\Pandas\Boris\worldstats.csv', index_col=['country', 'year'])
world.head()
```

```python
world_1 = world.stack()
```

```python
world_1.to_frame()
```

### The Unstack method

```python
world_1.unstack().unstack().head(5)
```

```python
world_1.unstack(level=2)
world_1.unstack(-1)
```

```python
world_1.unstack([1,0])
```

```python
world_1.unstack('year', fill_value=0).head(7)
```

### Pivot_table
