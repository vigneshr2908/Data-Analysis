```python
x1 = np.array([1, 3, 7, 11, 20])
y1 = np.array([5, 6, 2, 14, 19])

plt.plot(x1, y1, marker = 'P');
```

```python
x1 = np.array([1, 3, 7, 11, 20])
y1 = np.array([5, 6, 2, 14, 19])

# ms = marker size
# mfc = marker face color
# mec = marker edge color
# for mfc & mec, we can use hexadecimal color values

plt.plot(x1, y1, marker = 'P', ms = 10, mfc = 'c', mec = 'r');
```

```python
# for mfc & mec, we can use hexadecimal color values

plt.plot(x1, y1, marker = 'P', ms = 10, mfc = '#4CAF50', mec = 'c');
```
